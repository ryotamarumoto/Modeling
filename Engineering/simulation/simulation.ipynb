```python
#Kc=2の時のブロック線図

B = 38.1*pow(10, -3)*60/(7910*2*np.pi)

Ka = 2
Kc = 2
KA = Ka*Kc
KM = 28.3*pow(10, -3)
Ra = 17.9
KV = 10*(360/350)/(2*np.pi)
K = KA * (KM/Ra) * (1/20) * KV

J1 = 1.06*pow(10, -6)
J2 = math.pi*30*7930*(60**4)*pow(10, -15)*pow(0.5, 5)
J = J1 + pow(1/20, 2) * J2

zeta = B/(2*pow((J*K), 0.5))
omega = pow((K/J), 0.5)
V = tf([omega**2], [1, 2*zeta*omega, omega**2])

print(V)

matlab.bode(V)
plt.show()
```
