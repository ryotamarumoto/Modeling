```python
plt.xlabel('Dimensionless Frequency u')
plt.ylabel('Amplitude Ratio X0/Xb')

plt.ylim(0, 20)

x_1 = np.linspace(0.0, 0.8, 25)
y_1 = (1-x_1**2)/(x_1**4-(2+0.2)*x_1**2+1)

x_2 = np.linspace(0.8, 1.25, 25)
y_2 = (1-x_2**2)/(x_2**4-(2+0.2)*x_2**2+1)

x_3 = np.linspace(1.25, 2.5, 25)
y_3 = (1-x_3**2)/(x_3**4-(2+0.2)*x_3**2+1)

plt.plot(x_1, abs(y_1), 'black')
plt.plot(x_2, abs(y_2), 'black')
plt.plot(x_3, abs(y_3), 'black')

plt.vlines(0.8, 0, 20, linestyles='dotted', color='black')
plt.vlines(1.25, 0, 20, linestyles='dotted', color='black')
plt.grid()
plt.savefig('graph.png')
plt.show()
```

```python
def amprb_p(μ):
    u = ((1+μ/2)+(μ+(μ**2)/4)**0.5)**0.5
    return u

u = amprb_p(0.03551)

print(u)
```

```python
def amprb_m(μ):
    u_ = ((1+μ/2)-(μ+(μ**2)/4)**0.5)**0.5
    return u_

u_ = amprb_m(0.03551)

print(u_)
```
